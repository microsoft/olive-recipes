# FLUX.2-klein-4B — AMD NPU (VitisAI) inference sample

This notebook runs the optimized FLUX.2-klein-4B text-to-image pipeline produced by the
`flux_vitisai_workflow` recipe. The transformer and VAE decoder run on the AMD NPU via the
VitisAI execution provider; the text encoder and VAE encoder run on CPU.

The optimized pipeline directory (`./model/flux_vitisai`) contains the per-component ONNX
models plus the tokenizer / scheduler / `model_index.json` assembled by the workflow.

```python
model_dir = "./model/flux_vitisai"

ExecutionProvider="VitisAIExecutionProvider"
```

```python
from winml import register_execution_providers
register_execution_providers(ExecutionProvider)
```

```python
import numpy as np
from PIL import Image

# load_flux_npu_pipeline is provided by the FLUX.2-klein-4B RyzenAI runtime package
# (installed as part of the recipe requirements). It wires the NPU-compiled
# transformer / VAE decoder and the CPU text encoder into a Diffusers-style pipeline.
from flux_runtime import load_flux_npu_pipeline

pipeline = load_flux_npu_pipeline(model_dir, execution_provider=ExecutionProvider)

prompt = "A serene mountain lake at sunrise, photorealistic, ultra detailed"
result = pipeline(
    prompt,
    num_inference_steps=28,
    height=1024,
    width=1024,
    guidance_scale=4.0,
    generator=np.random.RandomState(seed=0),
)

from IPython.display import display
display(result.images[0])
```
